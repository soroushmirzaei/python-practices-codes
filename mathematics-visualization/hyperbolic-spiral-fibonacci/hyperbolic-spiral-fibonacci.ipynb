```python
#import mathematics statics libraries
import numpy as np
import cmath as cm

#import visualization libraries
import matplotlib.pyplot as plt
```

```python
#define equations
def phi(frac = 5):
    return 0.5 * (1 + np.sqrt(frac))

#define spiarl fibonacci equations
def spiral_fib(phi, x_start_point, x_end_point, y_start_point, y_end_point):
    x = np.arange(x_start_point, x_end_point, 0.01)
    y = np.arange(y_start_point, y_end_point, 0.01)
    x, y = np.meshgrid(x, y)
    part_a = ((np.power(phi, x) + np.power(phi, -x)) / np.sqrt(5)) - y
    part_b = ((np.power(phi, x) - np.power(phi, -x)) / np.sqrt(5)) + y
    square_z = part_a * part_b
    z = np.sqrt(square_z, dtype = complex)

    return x, y, np.real(z), np.imag(z)
```

```python
#plot equation
def plot_eq(x, y, z):
    fig = plt.figure(figsize = (12, 9))
    ax = plt.axes(projection = '3d')
    
    #plot
    ax.plot_surface(x, y, z, cmap = 'viridis')

    ax.set_xlabel('X Axis')
    ax.set_ylabel('Y Axis')
    ax.set_zlabel('Z Axis')
    ax.set_title('Spiral Fibonacci')
    return plt.show()
```

```python
#fractional phi 2
x, y, real_z, imag_z = spiral_fib(phi(frac = 2), -10, 10, -5, 5)
plot_eq(x, y, real_z)
```

```python
#fractional phi 3
x, y, real_z, imag_z = spiral_fib(phi(frac = 3), -5, 10, -20, 20)
plot_eq(x, y, real_z)
```

```python
#fractional phi 5 (spiral fibonacci)
x, y, real_z, imag_z = spiral_fib(phi(frac = 5), -5, 10, -50, 50)
plot_eq(x, y, real_z)
```
