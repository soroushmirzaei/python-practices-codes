```python
#import mathematics statics libraries
import numpy as np
from numpy import cos, sin, tan, cosh, sinh, tanh
from numpy import exp, log, log2, log10, log1p
from numpy import add, subtract, divide, multiply

#import visualization libraries
import matplotlib.pyplot as plt
```

```python
#define equations
def trig_func(equation, x_start_point, x_end_point, y_start_point, y_end_point):
    x = np.arange(x_start_point, x_end_point, 0.001)
    y = np.arange(y_start_point, y_end_point, 0.001)
    x, y = np.meshgrid(x, y)
    z = x**2 + y**2
    z = globals()[equation](z)

    return x, y, z
```

```python
#plot equation
def plot_eq(x, y, z):
    fig = plt.figure(figsize = (12, 9))
    ax = plt.axes(projection = '3d')
    
    #plot
    ax.plot_surface(x, y, z, cmap = 'viridis')

    ax.set_xlabel('X Axis')
    ax.set_ylabel('Y Axis')
    ax.set_zlabel('Z Axis')
    ax.set_title('Function')
    return plt.show()
```

```python
#sine
x, y, z = trig_func('sin', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#cosine
x, y, z = trig_func('cos', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#tangent
x, y, z = trig_func('tan', -1, 1, -0.5, 0.5)
plot_eq(x, y, z)
```

```python
#sine hyper
x, y, z = trig_func('sinh', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#cosine hyper
x, y, z = trig_func('cosh', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#cosine hyper
x, y, z = trig_func('tanh', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#exp
x, y, z = trig_func('exp', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#log
x, y, z = trig_func('log', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#log2
x, y, z = trig_func('log2', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#define equations
def comp_trig_func(equation_frst, equation_scnd, operator,
                   x_start_point, x_end_point, y_start_point, y_end_point):
    x = np.arange(x_start_point, x_end_point, 0.001)
    y = np.arange(y_start_point, y_end_point, 0.001)
    x, y = np.meshgrid(x, y)
    z = x**2 + y**2

    if equation_scnd != None:
        z =  globals()[operator](globals()[equation_frst](z), globals()[equation_scnd](z))
    else:
        z =  globals()[operator](globals()[equation_frst](z), z)

    return x, y, z
```

```python
#sine - cosine
x, y, z = comp_trig_func('sin', 'cos', 'subtract', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#sine + cosine
x, y, z = comp_trig_func('sin', 'cos', 'add', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#sine * cosine
x, y, z = comp_trig_func('sin', 'cos', 'multiply', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#cosine - sine
x, y, z = comp_trig_func('cos', 'sin', 'subtract', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#sine / x
x, y, z = comp_trig_func('sin', None, 'divide', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#sine * x
x, y, z = comp_trig_func('sin', None, 'multiply', -2, 2, -2, 2)
plot_eq(x, y, z)
```

```python
#cosine * x
x, y, z = comp_trig_func('cos', None, 'multiply', -2, 2, -2, 2)
plot_eq(x, y, z)
```
