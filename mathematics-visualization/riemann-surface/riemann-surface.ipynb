```python
#import mathematics statics libraries
import numpy as np

#import visualization libraries
import matplotlib.pyplot as plt
```

```python
def riemann_surf(start_point, end_point):
    real = np.linspace(start_point, end_point, 1000)
    imag = np.linspace(start_point, end_point, 1000)
    real, imag = np.meshgrid(real, imag)
    radius = np.sqrt(real + imag*1j)
    return real, imag, np.real(radius), np.imag(radius)
    
```

```python
def plot_surf(x, y, z, xlabel = 'X Axis', ylabel = 'Y Axis', zlabel = 'Z Axis'):
    fig = plt.figure(figsize = (12, 9))
    ax = plt.axes(projection = '3d')

    #plot
    ax.plot_surface(x, y, z, cmap = 'viridis')

    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_zlabel(zlabel)
    ax.set_title('Riemann Surface')
    return plt.show()
```

```python
real, imag, z_real, z_imag = riemann_surf(-1, 1)
plot_surf(imag, real, z_real, xlabel = 'Imag', ylabel = 'Real', zlabel = 'Z^(1/2) Real')
```

```python
real, imag, z_real, z_imag = riemann_surf(-1, 1)
plot_surf(real, imag, z_imag, xlabel = 'Real', ylabel = 'Imag', zlabel = 'Z^(1/2) Imag')
```

```python
real, imag, z_real, z_imag = riemann_surf(-1, 1)
plot_surf(z_imag, real, z_real, xlabel = 'Z^(1/2) Imag', ylabel = 'Real', zlabel = 'Z^(1/2) Real')
```

```python
real, imag, z_real, z_imag = riemann_surf(-1, 1)
plot_surf(np.concatenate([imag, imag]), np.concatenate([real, real]), np.concatenate([z_real ,-1 * z_real]),
          xlabel = 'Imag', ylabel = 'Real', zlabel = 'Z^(1/2) Real')
```
